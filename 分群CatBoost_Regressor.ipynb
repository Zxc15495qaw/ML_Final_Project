```python
# 定義連續型&離散型特徵
continuous_cols = ['Age', 'Annual Income', 'Number of Dependents', 'Health Score', 'Vehicle Age', 'Credit Score', 'Insurance Duration']
discrete_cols = ['Gender', 'Marital Status', 'Education Level', 'Occupation', 
                 'Location', 'Policy Type', 'Previous Claims', 'Customer Feedback', 
                 'Smoking Status', 'Exercise Frequency', 'Property Type']
```

```python
# 訓練資料與測試資料合併，便於處理
data_combined = pd.concat([train_data, test_data], keys=['train', 'test'])

# 使用中位數填補
imputer_continuous = SimpleImputer(strategy='median')
data_combined[continuous_cols] = imputer_continuous.fit_transform(data_combined[continuous_cols])

# 離散型特徵處理（將缺失值設為新類別 "Missing"）
# 將缺失值填補為 'Missing'
for col in discrete_cols:
    data_combined[col] = data_combined[col].fillna('Missing')


# 提取年份
data_combined['Policy Start Date'] = pd.to_datetime(data_combined['Policy Start Date'])
data_combined['Policy Start Year'] = data_combined['Policy Start Date'].dt.year
continuous_cols.append('Policy Start Year')

# 提取月份
data_combined['Policy Start Date'] = pd.to_datetime(data_combined['Policy Start Date'])
data_combined['Policy Start Month'] = data_combined['Policy Start Date'].dt.month
continuous_cols.append('Policy Start Month')

# 轉換日期並提取月初、月中、月底分類
data_combined['Policy Start Date'] = pd.to_datetime(data_combined['Policy Start Date'])
data_combined['Policy Start Period'] = data_combined['Policy Start Date'].dt.day.apply(
    lambda day: 'Early Month' if day <= 10 else 'Mid Month' if day <= 20 else 'Late Month'
)
discrete_cols.append('Policy Start Period')
```

```python
# 分離訓練資料與測試資料
train_data_cleaned = data_combined.xs('train')
test_data_cleaned = data_combined.xs('test')

# 確認填補情況
train_missing = train_data_cleaned.isnull().sum()
test_missing = test_data_cleaned.isnull().sum()

# 合併兩者，便於視覺化比較
missing_data = pd.DataFrame({
    'Train Missing': train_missing,
    'Test Missing': test_missing
}).sort_values(by='Train Missing', ascending=False)

# 可視化缺失值情況
plt.figure(figsize=(10, 6))
missing_data.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Missing Data Comparison: Train vs Test')
plt.xlabel('Features')
plt.ylabel('Number of Missing Values')
plt.xticks(rotation=90)
plt.show()
```

## K-Means

```python
# 提取連續型特徵
X_train_continuous = X_train[continuous_cols]
scaler = Normalizer()
X_train_continuous = scaler.fit_transform(X_train_continuous)
X_train_continuous = pd.DataFrame(X_train_continuous, columns=continuous_cols)

# 使用 K-Means 聚類進行不同群數的比較
costs = []
K = range(2, 11)  # 假設考慮 2 到 10 群
for k in K:
    kmeans = KMeans(n_clusters=k, n_init=5, random_state=42)
    kmeans.fit(X_train_continuous)  # 只用連續型特徵進行聚類
    costs.append(kmeans.inertia_)  # inertia_ 代表每個群內的總變異

# 繪製肘部法圖形
plt.figure(figsize=(8, 5))
plt.plot(K, costs, 'bx-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia (Cost)')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
X_train_discrete = X_train[discrete_cols]
X_train_discrete_encoded = pd.get_dummies(X_train_discrete, drop_first=True)

# 合併連續型特徵和編碼後的離散型特徵
X_train_continuous = X_train_continuous.reset_index(drop=True)
X_train_discrete_encoded = X_train_discrete_encoded.reset_index(drop=True)

# 合併資料
X_train_processed = pd.concat([X_train_continuous, X_train_discrete_encoded], axis=1)
```

```python
num_groups = 4
kmeans = KMeans(n_clusters=num_groups, n_init=10, random_state=42)
kmeans.fit(X_train_processed[continuous_cols])
labels = kmeans.labels_+ 1
```

```python
models = {}
def train_model_for_group(i):
    try:
        X_group = X_train_processed[labels == i]
        y_group = y_train[labels == i]
        model = CatBoostRegressor(iterations=500, learning_rate=0.1, verbose=0)
        model.fit(X_group, y_group)
        return i, model
    except Exception as e:
        print(f"Error in group {i}: {e}")
        raise

with ThreadPoolExecutor(max_workers=8) as executor:
    futures = [executor.submit(train_model_for_group, i) for i in range(1, num_groups + 1)]
    for future in futures:
        group, model = future.result()
        models[group] = model
```

```python
unique_labels, counts = np.unique(labels, return_counts=True)

# 輸出每個群組的數量
for label, count in zip(unique_labels, counts):
    print(f"Group {label}: {count} samples")
```

```python
# 計算預測值並計算殘差
residuals = {}
for i, model in models.items():
    X_group = X_train_processed[labels == i]
    y_group = y_train[labels == i]
    y_pred = model.predict(X_group)  # 計算預測值
    residuals[i] = y_group - y_pred  # 計算殘差
    # 計算 MAPE
    mape = mean_absolute_percentage_error(np.exp(y_group), np.exp(y_pred))
    
    # 輸出結果
    print(f"Group {i} - MAPE: {mape:.4f}")
```

```python
# 設定隨機抽樣的數量
sample_size = 1000  # 每個群組隨機抽取1000個點

# 創建 2x2 的子圖布局，將每個群組的殘差繪製在不同的子圖中
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# 設定群組標籤
group_labels = ['Group 1', 'Group 2', 'Group 3', 'Group 4']

# 迭代每個群組，將其殘差繪製在對應的子圖中
for i, ax in enumerate(axes.flatten()):
    # 確保 residuals[i] 是 numpy 陣列，再進行隨機抽樣
    residuals_array = np.array(residuals.get(i + 1, []))
    
    # 隨機抽樣
    sample_indices = np.random.choice(len(residuals_array), sample_size, replace=False)
    sampled_residuals = residuals_array[sample_indices]
    
    ax.scatter(range(sample_size), sampled_residuals, alpha=0.6)
    ax.axhline(y=0, color='r', linestyle='--', label="Zero Line")
    ax.set_title(f"Residual Plot - {group_labels[i]}")
    ax.set_xlabel("Index")
    ax.set_ylabel("Residuals")
    ax.legend(loc='upper right')  # 手動設置圖例位置

# 顯示所有圖
plt.tight_layout()
plt.show()
```

```python
weights = {}
weight_factor = 0.6  # 這是調整權重的因子，可根據需要調整
for i, residual in residuals.items():
    abs_residuals = np.abs(residual)  # 計算每個資料點的絕對殘差
    weight = 1 / (1 + weight_factor * abs_residuals)  # 計算每個資料點的權重
    weights[i] = weight  # 儲存每個群組的權重
```

```python
for i, model in models.items():
    # 取出該群組的訓練資料
    X_group = X_train_processed[labels == i]
    y_group = y_train[labels == i]
    
    # 取出對應的權重
    weight = weights[i]
    
    # 訓練模型
    model.fit(X_group, y_group, sample_weight=weight)
    
    # 預測
    y_pred = model.predict(X_group)

    # 計算 MAPE
    mape = mean_absolute_percentage_error(np.exp(y_group), np.exp(y_pred))
    
    # 輸出結果
    print(f"Group {i} - MAPE: {mape:.4f}")
```

```python
# 提取連續型特徵
X_val_continuous = X_val[continuous_cols]
scaler = Normalizer()
X_val_continuous = scaler.fit_transform(X_val_continuous)
X_val_continuous = pd.DataFrame(X_val_continuous, columns=continuous_cols)

# 確保驗證集的離散型特徵欄位與訓練集一致，並移除訓練集中不存在的欄位
X_val_discrete = X_val[discrete_cols]
X_val_discrete_encoded = pd.get_dummies(X_val_discrete, drop_first=True)
X_val_discrete_encoded = X_val_discrete_encoded.reindex(columns=X_train_discrete_encoded.columns, fill_value=0)

# 合併連續型特徵和編碼後的離散型特徵
X_val_continuous = X_val_continuous.reset_index(drop=True)
X_val_discrete_encoded = X_val_discrete_encoded.reset_index(drop=True)
X_val_processed = pd.concat([X_val_continuous, X_val_discrete_encoded], axis=1)
```

```python
# 使用 KMeans 模型對驗證集進行分群
labels_val = kmeans.predict(X_val_processed[continuous_cols])+1

# 預測每個群組的 y 值
for i, model in models.items():
    # 取出該群組的驗證集資料
    X_group_val = X_val_processed[labels_val == i]  # 使用 kmeans 在驗證集上進行分群後的 labels_val
    y_group_val = y_val[labels_val == i]
    
    # 預測
    y_pred_val = model.predict(X_group_val)
    
    # 計算 MAPE
    mape = mean_absolute_percentage_error(np.exp(y_group_val), np.exp(y_pred_val))
    
    # 輸出結果
    print(f"Group {i} - MAPE: {mape:.4f}")
```

### test data

```python
# 提取連續型特徵
X_test_continuous = X_test[continuous_cols]
scaler = Normalizer()
X_test_continuous = scaler.fit_transform(X_test_continuous)
X_test_continuous = pd.DataFrame(X_test_continuous, columns=continuous_cols)

# 確保驗證集的離散型特徵欄位與訓練集一致，並移除訓練集中不存在的欄位
X_test_discrete = X_test[discrete_cols]
X_test_discrete_encoded = pd.get_dummies(X_test_discrete, drop_first=True)
X_test_discrete_encoded = X_test_discrete_encoded.reindex(columns=X_train_discrete_encoded.columns, fill_value=0)

# 合併連續型特徵和編碼後的離散型特徵
X_test_continuous = X_test_continuous.reset_index(drop=True)
X_test_discrete_encoded = X_test_discrete_encoded.reset_index(drop=True)
X_test_processed = pd.concat([X_test_continuous, X_test_discrete_encoded], axis=1)
```

```python
# 使用 KMeans 模型對驗證集進行分群
labels_test = kmeans.predict(X_test_processed[continuous_cols])+1

# 初始化空的 DataFrame 用於儲存預測結果
predictions_test_df = pd.DataFrame()

# 假設 X_test 中有 'id' 欄位，這裡用來保存每個樣本的 id
id_column = X_test['id']  # 請根據實際的列名修改 'id' 欄位名稱

# 針對每個群組，預測並儲存結果
for i, model in models.items():
    # 取出該群組的測試集資料
    X_group_test = X_test_processed[labels_test == i]
    ids_group_test = id_column[labels_test == i]  # 取出對應群組的 id
    
    # 預測
    y_pred_test = np.exp(model.predict(X_group_test))
    
    # 將 id 和預測結果儲存到 predictions_test_df
    group_predictions_df = pd.DataFrame({
        'id': ids_group_test,
        'prediction': y_pred_test
    })
    
    # 合併到總的預測結果 DataFrame 中
    predictions_test_df = pd.concat([predictions_test_df, group_predictions_df], axis=0)


```

```python
# 重置索引，並確保資料整齊
predictions_test_df = predictions_test_df.reset_index(drop=True)

# 依照 'id' 排列
predictions_test_df = predictions_test_df.sort_values(by='id')

# 輸出最終的預測結果
print(predictions_test_df)
```
